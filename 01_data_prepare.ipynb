```python
root='/content/drive/Shareddrives/専門演習I/後期後半/SpeechData'#@param{type:'string'}
```

```python
import os, sys
import glob
import pandas as pd
#import torchaudio
#import torch

valid=['11','12','13','14','15']
test=['16','17','18','19','20']

data={'path':[], 'data_type':[], 'speaker': [] }
for path in glob.glob(os.path.join(root, '*/*wav')):
    #print(path)
    speaker = os.path.basename(os.path.dirname(path))
    if '_' not in os.path.basename(path):
      #data['data_type'].append('train')
      print(path)
      continue
    else:
      #print(path)
      num=os.path.splitext(os.path.basename(path))[0].split('_')[1]
      #if num not in test:
        #wave, sr = torchaudio.load(os.path.join(root, path))
        #if wave.shape[-1] > sr * 5:
          #print(path)
          #continue
      if num in valid:
          data['data_type'].append('valid')
      elif num in test:
          data['data_type'].append('eval')
      else:
          data['data_type'].append('train')
      data['speaker'].append(speaker)
      data['path'].append(os.path.join(root, path))

print(len(data['data_type']) , len(data['path']))
if len(data['data_type']) == len(data['path']):
  df = pd.DataFrame.from_dict(data, orient='columns')
  df.to_csv(os.path.join(root, 'data.csv'), index=False)
```
